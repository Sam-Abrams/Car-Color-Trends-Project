```python
## Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from summarytools import dfSummary
```

```python
## Import Data from cleaned cars.com dataset

df_cars_com = pd.read_csv('/Users/samabrams/Data Analysis Projects/Car-Color-Trends-Project/02 Data/Prepared Data/Cars.com Cleaned.csv')
```

```python
## Data Check

df_cars_com.head()
```

```python
## Import Data from cleaned auction dataset

df_cars_auction = pd.read_csv('/Users/samabrams/Data Analysis Projects/Car-Color-Trends-Project/02 Data/Prepared Data/df_cars_cleaned_final.csv')
```

```python
df_cars_auction.head()
```

```python
## Get column names for merging

print(df_cars_com.columns.tolist())
print(df_cars_auction.columns.tolist())
```

```python
## Time to match these up. First, I'm going to drop columns that I don't need from the auction dataset.

df_cars_auction = df_cars_auction.drop(columns=['seller', 'mmr', 'saledate', 'condition', 'trim'])
df_cars_auction.shape
```

```python
## Now to rename the cars.com columns to match the auction dataset

df_cars_com = df_cars_com.rename(columns={
    'brand': 'make',
    'model_year': 'year',
    'milage': 'odometer',
    'Color': 'color',
    'price': 'sellingprice'
    })

df_cars_com.head()
```

```python
## Now I need to add columns to the cars.com dataset that are in the auction dataset but not in the cars.com dataset

df_cars_com['vin'] = 'Unknown'
df_cars_com['state'] = 'Unknown'
df_cars_com['body'] = 'Unknown'

## And I need to add fuel_type to the auction dataset
df_cars_auction['fuel_type'] = 'Unknown'

## I want to add a source column to both datasets to indicate where the data came from.

df_cars_com['source'] = 'cars.com'
df_cars_auction['source'] = 'auction'
```

```python
## Check both datasets have matching columns

print("Cars.com columns:", sorted(df_cars_com.columns.tolist()))
print("Auction columns:", sorted(df_cars_auction.columns.tolist()))
```

```python
## Now I can merge the two datasets together

df_cars = pd.concat([df_cars_com, df_cars_auction], ignore_index=True)

## Check the merged dataset

df_cars.head()
```

```python
print(df_cars_auction.shape)
print(df_cars_com.shape)
print(df_cars.shape)
```

```python
## Check for duplicates in the merged dataset

duplicates = df_cars.duplicated(subset=['vin', 'make', 'model', 'year', 'color', 'odometer', 'sellingprice', 'fuel_type', 'state', 'body'])
print("Number of duplicates:", duplicates.sum())
```

```python
## Check for missing values in the merged dataset

missing_values = df_cars.isnull().sum()
print("Missing values in each column:")
print(missing_values[missing_values > 0])
```

```python
## Let's investigate the missing transmission values...

df_cars[df_cars['transmission'].isnull()]
```

```python
## Now it makes sense, pandas was reading NaN as missing values. I'm going to change those to 'Unknown' for consistency.

df_cars['transmission'] = df_cars['transmission'].fillna('Unknown')
df_cars['transmission'].value_counts()
```

```python
## And to confirm that the missing values have been filled...

print(df_cars['transmission'].isnull().sum())
```

```python
## Looking good! Now I can save the merged dataset to a new CSV file.

df_cars_combined = df_cars.to_csv('/Users/samabrams/Data Analysis Projects/Car-Color-Trends-Project/02 Data/Prepared Data/df_cars_combined.csv', index=False)
```
